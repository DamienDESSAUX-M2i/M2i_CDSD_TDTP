```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

import numpy as np
```

```python
X, y = load_breast_cancer(return_X_y=True)
print(f"Dataset : {X.shape}")
print(f"Classes : {np.unique(y)}")
print(f"proportion : {np.bincount(y) / len(y)}")
```

```python
test_size = 0.3

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, random_state=42
)

print(f"train : {X_train.shape}, test : {X_test.shape}")
print(f"proportion train : {np.bincount(y_train) / len(y_train)}")
print(f"proportion test : {np.bincount(y_test) / len(y_test)}")
```

```python
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X, y, test_size=test_size, random_state=42, stratify=y
)

print(f"train : {X_train_s.shape}, test : {X_test_s.shape}")
print(f"proportion train : {np.bincount(y_train_s) / len(y_train_s)}")
print(f"proportion test : {np.bincount(y_test_s) / len(y_test_s)}")
```

```python
bins = len(np.unique(y))

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(y_train, bins=bins, alpha=0.7, label="train")
axes[0].hist(y_test, bins=bins, alpha=0.7, label="test")
axes[0].set_title("split simple")
axes[0].set_xlabel("classe")
axes[0].set_ylabel("enchantillons")
axes[0].legend()

axes[1].hist(y_train_s, bins=bins, alpha=0.7, label="train")
axes[1].hist(y_test_s, bins=bins, alpha=0.7, label="test")
axes[1].set_title("split stratifie")
axes[1].set_xlabel("classe")
axes[1].set_ylabel("enchantillons")
axes[1].legend()

plt.tight_layout()
plt.show()
```
