```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
```

```python
tweets = [
    "Just finished my breakfast!! Check out https://t.co/xyz #goodMorning ☀️😊",
    "Loving the new #Python3.10 features... performance is amazing!!!",
    "@john_doe The movie was REALLY bad... waste of time :(",
    "Don't miss our special offer!!! Visit our website: https://example.com",
    "I can't believe it's Friday!!! #TGIF #Weekend",
    "Machine Learning & Deep Learning are the future of tech!!! 🚀🚀🚀"
]
```

### Tâche 1 : Conversion et gestion de base
1. Convertissez tous les tweets en minuscules
2. Supprimez les URLs (identifiez-les avec une regex)
3. Supprimez les mentions (@user) et les hashtags
4. Supprimez les emojis
5. Remplacez les caractères dupliqués excessifs (!!!, ???, ...) par un seul exemplaire

```python
def clean_text_basic(text):
    # 1. Convertir en minuscules
    text = text.lower()
    
    # 2. Supprimer les URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # 3. Supprimer les mentions (@user) et hashtags
    text = re.sub(r'@\w+|#\w+', '', text)
    
    # 4. Supprimer les emojis
    text = re.sub(r'[^\w\s\']', '', text)
    
    # 5. Remplacer les caractères dupliqués excessifs
    text = re.sub(r'(!)\1{1,}|\?\?+|\.{2,}', lambda m: m.group(0)[0], text)

    return text

cleaned_basic = [clean_text_basic(tweet) for tweet in tweets]

for tweet in cleaned_basic:
    print(tweet)
```

### Tâche 2 : Gestion de la ponctuation et des caractères spéciaux
1. Supprimez la ponctuation inutile, SAUF les apostrophes qui doivent être conservées pour les contractions
2. Supprimez les espaces multiples
3. Gérez les cas particuliers :
   - "can't" doit rester "can't" (contraction)
   - "don't" doit rester "don't"
4. Testez votre logique sur : `"I can't believe... it's AMAZING!!!"`

```python
def clean_punctuation(text):
    # Remplacer les espaces multiples par un seul
    text = re.sub(r'\s+', ' ', text)
    
    # Supprimer la ponctuation sauf les apostrophes
    text = re.sub(r"[^\w\s']", '', text)
    
    # Trim
    text = text.strip()
    
    return text

text_test = "I can't believe... it's AMAZING!!!"
print(clean_punctuation(text_test))

cleaned_punct = [clean_punctuation(tweet) for tweet in cleaned_basic]

for tweet in cleaned_punct:
    print(tweet)
```

### Tâche 3 : Tokenization et stopwords
1. Tokenisez le texte nettoyé (utilisez NLTK ou une approche simple split)
2. Créez une liste personnalisée de stopwords en français ET en anglais
3. Supprimez les stopwords
4. Gardez les contractions comme "can't", "don't" comme tokens entiers

```python
nltk.download("wordnet")
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('omw-1.4')
stop_words_en = set(stopwords.words('english'))
stop_words_fr = set(stopwords.words('french'))
all_stopwords = stop_words_en.union(stop_words_fr)

def tokenize_and_remove_stopwords(text):
    # Tokenize
    # tokens = word_tokenize(text)
    tokens = text.split()
    
    # Filtrer les stopwords
    filtered_tokens = [token for token in tokens if token.lower() not in all_stopwords]
    
    return filtered_tokens

tokenized = [tokenize_and_remove_stopwords(tweet) for tweet in cleaned_punct]

for tweet in tokenized:
    print(tweet)
```

### Tâche 4 : Lemmatisation
1. Implémentez la lemmatisation avec NLTK WordNetLemmatizer
2. Appliquez-la sur les tokens filtrés
3. Comparez avant/après (montrez des exemples)
4. Optionnel : Utilisez SpaCy pour un résultat amélioré


```python
lemmatizer = WordNetLemmatizer()

def lemmatize_tokens(tokens):
    lemmatized = [lemmatizer.lemmatize(token, pos="v") for token in tokens]
    return lemmatized

# Appliquez la lemmatisation
lemmatized = [lemmatize_tokens(tokens) for tokens in tokenized]

print("Après lemmatisation:")
for i, tokens in enumerate(lemmatized, 1):
    print(f"{i}. {tokens}")

# Comparaison avant/après
print("\nExemples de lemmatisation:")
examples = ["learning", "loved", "amazing", "believing"]
for word in examples:
    lemma = lemmatizer.lemmatize(word, pos="v")
    print(f"{word} -> {lemma}")
```

### Tâche 5 : Pipeline complet
1. Créez une fonction `clean_tweet(text)` qui applique TOUTES les étapes précédentes
2. Testez-la sur l'ensemble du corpus
3. Affichez avant/après pour 3 tweets
4. Mesurez la réduction du vocabulaire (nombre de tokens uniques avant/après)

```python
def clean_tweet_complete(text):
    # Nettoyage basique
    text = clean_text_basic(text)
    
    # Gestion de la ponctuation
    text = clean_punctuation(text)
    
    # Tokenization et suppression stopwords
    tokens = tokenize_and_remove_stopwords(text)
    
    # Lemmatisation
    tokens = lemmatize_tokens(tokens)
    
    return tokens

final = [clean_tweet_complete(tweet) for tweet in tweets]
print(final)
```

- Gérez les cas spéciaux français : accents, traits d'union

```python
import unicodedata

def clean_french_text(text):
    # Option 1 : Normalisation (conserver les accents)
    text = unicodedata.normalize('NFC', text)
    
    # Option 2 : Gestion des traits d'union (les conserver pour les mots composés)
    # On ne supprime pas les traits d'union entre lettres
    text = re.sub(r'(?<=[a-zàâäéèêëïîôùûüç])-(?=[a-zàâäéèêëïîôùûüç])', ' ', text, flags=re.IGNORECASE)
    
    return text

def remove_accents(text):
    # Normalisation NFD : décompose les caractères accentués
    nfd = unicodedata.normalize('NFD', text)
    # Supprime les marques diacritiques
    return ''.join(char for char in nfd if unicodedata.category(char) != 'Mn')

# Exemples avec du texte français
french_texts = [
    "L'Intelligence Artificielle est très-importante aujourd'hui",
    "Café, thé et protéine à jeun : c'est génial !",
    "Jean-Pierre a mangé des crêpes à côté de l'école",
    "État-nation, arc-en-ciel, rendez-vous"
]

print("TRAITEMENT DU TEXTE FRANÇAIS")
print("="*80)

for text in french_texts:
    print(f"\nOriginal:          {text}")
    print(f"Avec accents:      {clean_french_text(text)}")
    print(f"Sans accents:      {remove_accents(text)}")
    
# Intégration dans le pipeline
def clean_tweet_with_french(text, remove_accent=False):
    if remove_accent:
        text = remove_accents(text)
    else:
        text = clean_french_text(text)
    
    # Reste du pipeline
    text = clean_text_basic(text)
    text = clean_punctuation(text)
    tokens = tokenize_and_remove_stopwords(text)
    tokens = lemmatize_tokens(tokens)
    
    return tokens

# Test avec texte français
test_fr = "L'IA et le deep-learning révolutionnent l'éducation française"
print(f"\n\nTest Pipeline Français:")
print(f"Original: {test_fr}")
print(f"Avec accents: {clean_tweet_with_french(test_fr, remove_accent=False)}")
print(f"Sans accents: {clean_tweet_with_french(test_fr, remove_accent=True)}")
```

- Créez une classe `TextCleaner` réutilisable avec configuration

```python
class TextCleaner:
    
    def __init__(self, 
                 lowercase=True, 
                 remove_urls=True,
                 remove_mentions=True,
                 remove_hashtags=True,
                 remove_emojis=True,
                 remove_accents=False,
                 remove_stopwords=True,
                 lemmatize=True,
                 stem=False,
                 language='english'):
        self.lowercase = lowercase
        self.remove_urls = remove_urls
        self.remove_mentions = remove_mentions
        self.remove_hashtags = remove_hashtags
        self.remove_emojis = remove_emojis
        self.remove_accents = remove_accents
        self.remove_stopwords = remove_stopwords
        self.lemmatize = lemmatize
        self.stem = stem
        self.language = language
        
        # Initialisation des outils
        if self.remove_stopwords:
            self.stopwords = set(stopwords.words(language))
        
        if self.lemmatize:
            self.lemmatizer = WordNetLemmatizer()
        
        if self.stem:
            self.stemmer = PorterStemmer()
    
    def clean(self, text):
        # Lowercase
        if self.lowercase:
            text = text.lower()
        
        # Supprimer accents
        if self.remove_accents:
            nfd = unicodedata.normalize('NFD', text)
            text = ''.join(char for char in nfd if unicodedata.category(char) != 'Mn')
        
        # URLs
        if self.remove_urls:
            text = re.sub(r'http\S+|www\S+', '', text)
        
        # Mentions
        if self.remove_mentions:
            text = re.sub(r'@\w+', '', text)
        
        # Hashtags
        if self.remove_hashtags:
            text = re.sub(r'#\w+', '', text)
        
        # Emojis et caractères spéciaux
        if self.remove_emojis:
            text = re.sub(r'[^\w\s\']', '', text)
        
        # Caractères dupliqués
        text = re.sub(r'(!)\1{1,}|\?\?+|\.{2,}', lambda m: m.group(0)[0], text)
        
        # Normaliser espaces
        text = re.sub(r'\s+', ' ', text).strip()
        
        # Tokenization
        tokens = word_tokenize(text)
        
        # Stopwords
        if self.remove_stopwords:
            tokens = [token for token in tokens if token.lower() not in self.stopwords]
        
        # Lemmatisation
        if self.lemmatize:
            tokens = [self.lemmatizer.lemmatize(token) for token in tokens]
        
        # Stemming
        if self.stem:
            tokens = [self.stemmer.stem(token) for token in tokens]
        
        return tokens
    
    def clean_corpus(self, corpus):
        return [self.clean(text) for text in corpus]
    
cleaner1 = TextCleaner(
    lowercase=True,
    remove_urls=False,
    remove_mentions=True,
    remove_hashtags=True,
    remove_emojis=True,
    remove_stopwords=True,
    lemmatize=True,
    stem=False,
    language='french'
)

for tweet in tweets:
    cleaned = cleaner1.clean(tweet)
    print(cleaned)
```
