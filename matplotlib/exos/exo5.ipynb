```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
```

# Exercice

1. **Subplot 1** : Ligne - Nombre de publications hebdomadaires
2. **Subplot 2** : Barres - Taux d'engagement par plateforme
3. **Subplot 3** : Pie - Répartition des abonnés par plateforme
4. **Subplot 4** : Scatter - Portée vs Interactions

```python
np.random.seed(888)

semaines = [f'S{i}' for i in range(1, 13)]
publications = np.random.randint(25, 80, len(semaines)) + np.linspace(0, 20, len(semaines))
df_publications = pd.DataFrame({'Semaine': semaines, 'Publications': publications.astype(int)})

plateformes = ['Instagram', 'Facebook', 'Twitter', 'TikTok', 'LinkedIn']
taux_engagement = [4.2, 2.8, 3.5, 6.1, 2.3]
df_engagement = pd.DataFrame({'Plateforme': plateformes, 'Taux_Engagement': taux_engagement})
df_engagement = df_engagement.sort_values('Taux_Engagement')

abonnes = [45000, 62000, 38000, 28000, 15000]
df_abonnes = pd.DataFrame({'Plateforme': plateformes, 'Abonnes': abonnes})

n_posts = 30
df_posts = pd.DataFrame({
    'Portee': np.random.uniform(1000, 50000, n_posts),
    'Interactions': np.random.uniform(100, 8000, n_posts),
    'Plateforme': np.random.choice(plateformes, n_posts)
})
```

```python
fig = make_subplots(
    rows=2,
    cols=2,
    subplot_titles=("Nombre de publications hebdomadaires", "Taux d'engagement par plateforme", "Répartition des abonnés par plateforme", "Portée vs Interactions"),
    specs=[
        [{"type": "scatter"}, {"type": "bar"}],
        [{"type": "pie"}, {"type": "scatter"}]
    ]
)

fig.add_trace(go.Scatter(x=df_publications["Semaine"], y=df_publications["Publications"], name="Publications", showlegend=False), row=1, col=1)
fig.add_trace(go.Bar(x=df_engagement["Plateforme"], y=df_engagement["Taux_Engagement"], showlegend=True, orientation="v", name="Taux d'engagement", legendgroup="g1"), row=1, col=2)
fig.add_trace(go.Pie(labels=df_abonnes["Plateforme"], values=df_abonnes["Abonnes"], showlegend=True, legendgroup="g1"), row=2, col=1)
fig.add_trace(go.Scatter(x=df_posts["Portee"], y=df_posts["Interactions"], name="Portee vs Interactions", mode="markers", showlegend=False), row=2, col=2)

fig.update_layout(height=600, width=1200, title_text="Exo5")
```
