```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
dates = pd.date_range("2024-01-01", periods=30, freq="D")
ventes = np.random.randint(100, 500, size=30) + np.linspace(0, 100, 30)

# figsize : dfinit la taille du graphique (largeur, longueur)
plt.figure(figsize=(12, 4))

# subplot : Cre une grille de subplots (nb_ligne, nb_colonne, indice)
plt.subplot(1, 2, 1)

# color : couleur de la ligne
# linestyle : style de la ligne ("-" ou "solid" pour une ligne continue)
#  linewidth : paisseur de la ligne
# marker : ajoute un marqueur pour chaque point
plt.plot(dates, ventes, color="steelblue", linestyle="solid", linewidth=2, marker="+")

plt.title("Evolution des ventes", fontsize=14, fontweight="bold")

plt.xlabel("Date")
plt.ylabel("Ventes")

# rotation des lables de 45 sur l'axe x
plt.xticks(rotation=45)

plt.grid(True, alpha=0.4)

plt.subplot(1, 2, 2)

categories = ["Produit A", "Produit B", "Produit C", "Produit D"]
valeurs = [320, 450, 280, 390]

plt.bar(categories, valeurs, color=["#FF6B6B", "#4ECDC4", "#45B7D1", "#FFA07A"])

plt.title("Ventes par produits", fontsize=14, fontweight="bold")

plt.ylabel("Ventes")

plt.tight_layout()
```
