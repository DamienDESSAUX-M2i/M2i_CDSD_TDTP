```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colormaps
```

```python
np.random.seed(99)
mois = ['Jan', 'Fév', 'Mar', 'Avr', 'Mai', 'Jun']
ventes_region_a = [120, 150, 180, 160, 200, 220]
ventes_region_b = [100, 130, 140, 170, 190, 210]
satisfaction_client = np.random.uniform(3.5, 5, 100)
```

```python
fig, axes = plt.subplots(2,2,figsize=(14,10))

fig.suptitle("Dashboard d'analyse mensuelle", fontsize=16, fontweight="bold")

x = np.arange(len(mois))
width = 0.35
axes[0,0].bar(x - width/2, ventes_region_a, width, label="Region A", color="green")
axes[0,0].bar(x + width/2, ventes_region_b, width, label="Region B", color="yellow")
axes[0,0].set_title("Ventes par région", fontweight="bold")
axes[0,0].set_xticklabels(mois)
axes[0,0].set_ylabel("Ventes")
axes[0,0].legend()

axes[0,1].hist(satisfaction_client, bins=20, color="purple", edgecolor="red")
axes[0,1].set_title("Distribution satisfaction client", fontweight="bold")
axes[0,1].set_xlabel("Note")
axes[0,1].set_ylabel("Nombre de client")

axes[1,0].plot(mois, ventes_region_a, marker="+", label="Réfion A", linewidth=2, color="brown")
axes[1,0].plot(mois, ventes_region_b, marker="+", label="Réfion B", linewidth=2, color="magenta")
axes[1,0].set_title("Distribution satisfaction client", fontweight="bold")
axes[1,0].set_xlabel("Note")
axes[1,0].set_ylabel("Nombre de client")

produit = ["Produit 1", "Produit 2", "Produit 3", "Produit 4"]
values = [35, 4, 15, 22]

axes[1,1].pie(values, labels=produit, colors=["blue", "blue", "blue", "blue"])
axes[1,0].set_title("Répartition des ventes", fontweight="bold")

```
