# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
# plt.scatter(X,Y) -> nuage de point
# plt.plot(X,Y) -> ligne
# plt.bar(X,Y) -> diagramme en bar
# plt.hist(X,Y) -> Histogramme
# plt.pie(X,Y) -> diagramme circulaire
```

```python
X = np.linspace(0,100,100)
Y = np.random.randint(10, 90, 100)
plt.scatter(X,Y,s=60, alpha=0.5, c='red', marker='P')
# s -> taille des points
# alpha -> opacité des points
# c -> couleur
# marker -> style des points
```

```python
X = np.random.random(5)
Y = [f"{y}" for y in range(5)]
plt.pie(x=X, labels=Y, autopct='%.2f%%')
```

```python
X_sorted = np.sort(X)[::-1]
plt.bar(height=X_sorted, x=Y)
```

```python
X = np.random.binomial(100,0.4,1000)
plt.hist(X,12)
```

```python
X = np.linspace(0,1,10)
Y = np.random.normal(0,1,10)
plt.plot(X,Y,marker='o', linestyle='--', color='red')
```

```python
X = np.linspace(0,1,10)
Y1 = np.random.poisson(20,10)
Y2 = np.random.binomial(50,0.5,10)
Y3 = np.random.geometric(0.6,10)
Y = [Y1, Y2, Y3]

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }

plt.figure(figsize=(10,6))
for k in range(len(Y)):
  plt.scatter(X,Y[k],label=f"Y{k+1}")
plt.legend()
plt.text(0.6, 10, r'text', fontdict=font)
plt.title("titre", fontdict=font)
plt.xlabel("x_label", fontdict=font)
plt.ylabel('y_label', fontdict=font)
plt.ylim(0,40)
plt.yticks(np.linspace(0,40,11))
plt.grid(axis='y')
plt.show()
```

# Seaborn

Seaborn est une surcouche de `matplotlib`, il permet de travailler avec les `dataframe` de `pandas`.

```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('prets.csv')
sns.scatterplot(data=data, x='revenu', y='taux')
```

```python
sns.scatterplot(data=data, x='revenu', y='taux', hue='type')
```

```python
sns.scatterplot(data=data, x='revenu', y='taux', size='duree')
```

On utilise les fonctions de personnalisation de `matplotlib` pour `seaborn`.

```python
plt.figure(figsize=(8,4)) # Taille du plot
plt.rcParams.update({'font.size': 12}) # Modifie la font de tous les graphiques (labels, titres, legende, ...)
sns.scatterplot(data=data, x='revenu', y='taux', hue='type')
plt.ylabel("Taux du prêt")
plt.xlabel('Revenu mensuel (€)')
plt.grid()
plt.xlim(500, 7500)
plt.legend(bbox_to_anchor=(1, 1.02)) # Fixe la position de la légende en dehors du graphique aux coordonnée précisées
plt.title("Taux du prêt en fonction du revenu mensuel\npar type de prêt contracté")
plt.show()
```

On peut agréger des données directement avec `seaborn`.
- `errorbar='ci'` permet d'afficher les intervales de confiances
- `estimator=np.mean` permet d'utiliser une fonction d'agrégation, ici la moyenne

Interval de confiance :
- Moyenne empirique (non biaisé $\mathbb{E}(\bar{x})=\mu$) de l'échantillon : $\bar{x} = \frac{1}{n}\sum_{k=1}^{n}x_k$
- Varience empirique non biaisé ($n-1$ pour que $\mathbb{E}(\bar{s}^2)=\sigma^2$) de l'échantillon : $\bar{s}^2 = \frac{1}{n - 1}\sum_{k=1}^{n}(x_k-\bar{x})^2$
- niveau de sécurité : $1-\alpha=1-0.05=0.95$
- $\phi$ fonction de répartition de la loi normale centrée réduite
- $t_{\alpha}$ tel que $\phi(t_{\alpha})=1-\frac{\alpha}{2}=0.975$
- La table de la loi normal centrée réduite donne $t_{\alpha}=1.96$
- Interval de confiance : $I=[\bar{x}-1.96\frac{\bar{s}}{\sqrt{n}},\bar{x}+1.96\frac{\bar{s}}{\sqrt{n}}]$

```python
plt.figure(figsize=(8,4))
sns.barplot(data=data, x='ville', y='remboursement', errorbar='ci', estimator=np.mean, hue='type')
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(data=data, x='duree', y='remboursement', estimator=np.mean, errorbar=None) # pour tracer des courbes
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=data, x='duree', y='remboursement') # pour des boîtes à moustaches
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data=data, x='taux', bins=40, hue='type') # pour des histogrammes, bins correspond au nombre d'intervalles
```

```python
plt.figure(figsize=(8,4))
sns.kdeplot(data=data, x='taux', hue='type')
```

```python
plt.figure(figsize=(8,4))
sns.kdeplot(data=data, x='taux', y='duree', hue='type') # pour un graphique à densité
```

```python
plt.figure(figsize=(8,4))
sns.pairplot(data=data[['revenu', 'duree', 'taux', 'remboursement', 'type']], hue='type', diag_kind="hist") # pour tracer l’ensemble des variables numériques deux à deux, sur un seul graphique
```

```python
sns.countplot(data=data, x='type')
```

`Seaborn` contient des datasets comme iris ou mpg.
mpg contient des informations sur près de 400 voitures relativement anciennes (années de production entre 1970 et 1982).
- mpg  : le nombre de miles per gallon ;
- cylinders  : le nombre de cylindres du modèle ;
- displacement  : volume (en litres) de la cylindrée du moteur ;
- horsepower  : la puissance en chevaux du moteur ;
- weight  : le poids du modèle ;
- acceleration  : le nombre de secondes pour passer de 0 à 60mph (~96 km/h) ;
- model_year  : l’année de production du modèle ;
- origin  : le continent d’origine de la voiture (Europe, Japon ou USA) ;
- name  : le nom du modèle.

```python
mpg = sns.load_dataset('mpg') # Import du dataset mpg
```

```python
origin_count = mpg['origin'].value_counts().reset_index()
display(origin_count)
plt.pie(labels=origin_count['origin'], x=origin_count['count'], autopct='%.1f%%')
```
