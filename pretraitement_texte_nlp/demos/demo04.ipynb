```python
# Sentiment analysis avec TextBlob (lexicon-based)
# pip install textblob
from textblob import TextBlob

texte = "Ce film est genial! J'ai adore."
blob = TextBlob(texte)

# Polarity: -1 (negatif) a +1 (positif)
print(f"Polarity: {blob.sentiment.polarity}")
print(f"Subjectivity: {blob.sentiment.subjectivity}")

# Classification simple
if blob.sentiment.polarity > 0.1:
    sentiment = "POSITIF"
elif blob.sentiment.polarity < -0.1:
    sentiment = "NEGATIF"
else:
    sentiment = "NEUTRE"

print(sentiment)
```

```python
# pip install textblob-fr
from textblob import TextBlob
from textblob_fr import PatternAnalyzer, PatternTagger

texte = "Ce film est genial! J'ai adore."

blob = TextBlob(texte, pos_tagger=PatternTagger(), analyzer=PatternAnalyzer())
polarity = blob.sentiment[0]

# Classification simple
if polarity > 0.1:
    sentiment = "POSITIF"
elif polarity < -0.1:
    sentiment = "NEGATIF"
else:
    sentiment = "NEUTRE"

print(sentiment)
print(polarity)
```

```python
# pip install transformers
from transformers import (
    AutoModelForSequenceClassification,
    AutoTokenizer,
    TextClassificationPipeline,
)

# Charger un modele BERT pre-trained francais
model_name = "distilbert-base-multilingual-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

# Pipeline for quick inference
classifier = TextClassificationPipeline(model=model, tokenizer=tokenizer)

# Prediction
result = classifier("Ce film est fantastique!")
print(result)  # [{'label': 'POSITIVE', 'score': 0.998}]
```
