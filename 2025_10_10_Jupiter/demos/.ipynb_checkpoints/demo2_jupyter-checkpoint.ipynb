````python
import numpy as np
import matplotlib.pyplot as plt
````

## 3. Création de tableaux NumPy

````python
# Tableau de 0 à 9
a = np.arange(10)
print("Tableau a : ", a)

# Opération vectoriel
c = a * 2 + 1
print("Tableau c (2*a+1) : ", c)
````

## 4. Graphique simple

````python
# Mois
mois = np.arange(12)

# Benefices correspondant à chaque mois
benefice = [5, 7, 6, 8, 10, 12, 11, 13, 15, 14, 16, 18]

# Création du graphique
plt.plot(mois, benefice, marker='o', linestyle='-', color='blue')
plt.title("Bénéfice mensuel sur l'année")
plt.xlabel("Mois")
plt.ylabel("Bénéfices (k€)")
plt.xticks(mois)
plt.grid(True)
plt.show
````
